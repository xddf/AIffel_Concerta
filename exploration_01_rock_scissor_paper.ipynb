### Resize

- 1-7의 코드를 가져옴
- glob.glob가 작동하지 않아서 위에서 처럼 glob메서드를 한 번만 사용하는 것으로 수정
- 구글드라이브는 경로에 root를 넣으니 작동하지 않아서 수정
- 이후 계속 에러가 뜨는 상태였는데, 리네임 코드를 여러번 돌리던 과정에서 문제가 있었던 것으로 추정된다..ㅠ

````python
from PIL import Image

def resize_images(img_path):                                    # 리사이즈 함수 생성
    images=glob(img_path + '/*.jpg')                            # glob으로 경로를 images 변수에 담기
    print(len(images), 'images to be resized.')                 # 이미지 파일의 갯수와 문구 출력
    target_size=(28,28)                                         # 28x28로 사이즈 변환
    for img in images: #[1:]:
        """에러가 뜨고 엄청 고생한 뒤에 보니 압축 파일 확장자가 jpg로 리네임 되어 있었다. 
        리네임 코드에 분명 글롭이 선행되어 있어서 zip파일이 리네임 된 것이 이상한데, 
        코드를 여러번 수정하며 돌리던 과정에서 발생한 사고 같다.
        우선 첫 번째 항목을 무시하고 실행시키기로 결정"""
    
        old_img=Image.open(img)                                 # 원본 이미지를 가져와 새 이미지로 바꾸고, 저장하는 반복문
        new_img=old_img.resize(target_size,Image.ANTIALIAS)
        new_img.save(img, 'JPEG')
    
    print(len(images), 'images resized.')
````

````python

def load_data(img_path, number_of_data=7861):  # 가위바위보 이미지 개수 총합에 주의하세요.
    # 가위 : 0, 바위 : 1, 보 : 2
    img_size=28
    color=3
    #이미지 데이터와 라벨(가위 : 0, 바위 : 1, 보 : 2) 데이터를 담을 행렬(matrix) 영역을 생성합니다.
    imgs=np.zeros(number_of_data*img_size*img_size*color,dtype=np.int32).reshape(number_of_data,img_size,img_size,color)
    labels=np.zeros(number_of_data,dtype=np.int32)

    idx=0
    for file in iglob(img_path+'scissor/*.jpg'):
        img = np.array(Image.open(file),dtype=np.int32)
        imgs[idx,:,:,:]=img    # 데이터 영역에 이미지 행렬을 복사
        labels[idx]=0   # 가위 : 0
        idx=idx+1

    for file in iglob(img_path+'rock/*.jpg'):
        img = np.array(Image.open(file),dtype=np.int32)
        imgs[idx,:,:,:]=img    # 데이터 영역에 이미지 행렬을 복사
        labels[idx]=1   # 바위 : 1
        idx=idx+1  
    
    for file in iglob(img_path+'paper/*.jpg'):
        img = np.array(Image.open(file),dtype=np.int32)
        imgs[idx,:,:,:]=img    # 데이터 영역에 이미지 행렬을 복사
        labels[idx]=2   # 보 : 2
        idx=idx+1
        
    print("이미지 개수는", idx,"입니다.")
    return imgs, labels

(img_data, lable) = load_data(path)
img_data_norm = img_data/255.0   # 입력은 0~1 사이의 값으로 정규화

print("img_data: {}".format(img_data.shape))
print("lable: {}".format(lable.shape))
````

````python
def choice_train_test_split(X, y, test_size=0.2, shuffle=True, random_num):
    
    test_num = int(X.shape[0] * test_size)
    train_num = X.shape[0] - test_num
    
    if shuffle:
        np.random.seed(random_state = random_num)
        train_idx = np.random.choice(X.shape[0], train_num, replace=False)

        test_idx = np.setdiff1d(range(X.shape[0]), train_idx)
        
        X_train = X[train_idx,:]
        X_test = X[test_idx]
        y_train = y[train_idx,]
        y_test = y[test_idx]    

    else:
        X_train = X[:train_num]
        X_test = X[train_num:]
        y_train = y[:train_num]
        y_test = y[train_num:]

    return X_train, X_test, y_train, y_test

# 출처: https://rfriend.tistory.com/519 [R, Python 분석과 프로그래밍의 친구 (by R Friend)]
````

### 학습에 필요한 텐서플로우 라이브러리 불러오기

````python
#!pip install tensorflow==2.4.1

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM, BatchNormalization
from tensorflow.keras.callbacks import TensorBoard
````

````python
# 과정 중 발생한 에러가 LMS와 버전 차이 때문인지 확인하기 위해 작성했었다.
print(tf.__version__)
print(keras.__version__)
````

### 학습 모델 카피

````python
# 익스 1-6의 코드를 카피. 
# 세부적인 내용이 궁금하지만.. 
#여기서 더 혼자 헤메었다가는 업로드 자체를 못 할 것 같아서 다음 기회로.

n_channel_1=16
n_channel_2=32
n_dense=64
n_train_epoch=20

model=keras.models.Sequential()
model.add(keras.layers.Conv2D(n_channel_1, (3,3), activation='relu', input_shape=(28,28,3))) #input shape를 데이터와 맞춰준다.
model.add(keras.layers.MaxPool2D(2,2))
model.add(keras.layers.Conv2D(n_channel_2, (3,3), activation='relu'))
model.add(keras.layers.MaxPooling2D((2,2)))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(n_dense, activation='relu'))
model.add(keras.layers.Dense(3, activation='softmax')) #최종 분류 할 class의 갯수를 수정

model.summary()
````

### 학습 시키기
- 컴파일 인자들의 내용이 궁금하면서도 앞으로가 조금 두렵다(..)

````python
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])

X_train_norm = X_train/255.0
model.fit(X_train_norm, y_train, epochs=20)
````

### 테스트

````python
X_test_norm=X_test/255.0
test_loss, test_accuracy = model.evaluate(X_test_norm, y_test, verbose=2)
print("test_loss: {} ".format(test_loss))
print("test_accuracy: {}".format(test_accuracy))
````
